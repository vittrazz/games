```python
group_1 = data[[col for col in data.columns if col.startswith('1')]]
group_2 = data[[col for col in data.columns if col.startswith('2')]]
group_3 = data[[col for col in data.columns if col.startswith('3')]]
```

```python
# Построчное сравнение значений
comparison_results = pd.DataFrame()

# for index, row in data.iterrows():
#     row_group_1 = row[group_1.columns].values
#     row_group_2 = row[group_2.columns].values
#     row_group_3 = row[group_3.columns].values

#     # Сравнение: группа 1 > группа 2
#     comparison_results.loc[index, 'group_1_vs_group_2'] = all(row_group_1 <= row_group_2)

#     # Сравнение: группа 1 > группа 3
#     comparison_results.loc[index, 'group_1_vs_group_3'] = all(row_group_1 <= row_group_3)

#     # Сравнение: группа 2 > группа 3
#     comparison_results.loc[index, 'group_2_vs_group_3'] = all(row_group_2 <= row_group_3)

# print(comparison_results)
```

```python
comparison_results['mean_group_1'] = round(group_1.mean(axis=1), 2)
comparison_results['mean_group_2'] = round(group_2.mean(axis=1), 2)
comparison_results['mean_group_3'] = round(group_3.mean(axis=1), 2)
```

```python
comparison_results
```

```python
comparison_results['%_mean_gr_1_and_gr_2'] = round(((comparison_results['mean_group_2'] - comparison_results['mean_group_1'])) / comparison_results['mean_group_1'] * 100, 2)
comparison_results['%_mean_gr_2_and_gr_3'] = round(((comparison_results['mean_group_3'] - comparison_results['mean_group_2'])) / comparison_results['mean_group_2'] * 100, 2)
comparison_results['%_mean_gr_1_and_gr_3'] = round(((comparison_results['mean_group_3'] - comparison_results['mean_group_1'])) / comparison_results['mean_group_1'] * 100, 2)
```

```python
groups = {
    'behavior_group_1': ['1.1', '1.5', '1.8'],
    'behavior_group_2': ['2.1', '2.5', '2.8'],
    'behavior_group_3': ['3.1', '3.5', '3.8']
}

for group_name, columns in groups.items():
    comparison_results[group_name] = round(data[columns].mean(axis=1), 2)
```

```python
comparison_results['%_behavior_gr_1_and_gr_2'] = round(((comparison_results['behavior_group_2'] - comparison_results['behavior_group_1'])) / comparison_results['behavior_group_1'] * 100, 2)
comparison_results['%_behavior_gr_1_and_gr_3'] = round(((comparison_results['behavior_group_3'] - comparison_results['behavior_group_1'])) / comparison_results['behavior_group_1'] * 100, 2)
comparison_results['%_behavior_gr_2_and_gr_3'] = round(((comparison_results['behavior_group_3'] - comparison_results['behavior_group_2'])) / comparison_results['behavior_group_2'] * 100, 2)
```
